```python
import pandas as pd
```

```python
df = pd.read_csv('Pittsburgh_Street_Trees_2020-01-28.csv', low_memory = False)
```

```python
df.head()
```

```python
df['common_name'].value_counts()
```

```python
df['condition'].value_counts()
```

```python
df['condition'] = df['condition'].astype('category')
df['condition'].cat.set_categories(['Excellent', 'Very Good', 'Good', 'Fair', 'Poor', 'Critical', 'Dead'], inplace = True)
df['condition'] = df['condition'].cat.as_ordered()
```

```python
df.pivot_table('common_name', 'neighborhood', 'condition', 'count').style.bar(axis = 1)
```

```python
df.groupby('neighborhood')['common_name'].agg(pd.Series.mode)
```
